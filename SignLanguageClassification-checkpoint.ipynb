```python
# Create training and testing arrays
train_set = np.array(train_data, dtype = 'float32')
test_set = np.array(test_data, dtype='float32')
```

```python
# Prepare the training and testing dataset
X_train = train_set[:, 1:] / 255
y_train = train_set[:, 0]

X_test = test_set[:, 1:] / 255
y_test = test_set[:,0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_validate, y_train, y_validate = train_test_split(X_train, y_train, test_size = 0.2, random_state = 42)
print("train set shape")
print(X_train.shape)
print(y_train.shape)
print("validation set shape")
print(X_validate.shape)
print(y_validate.shape)
print("test set shape")
print(X_test.shape)
print(y_test.shape)
```


#**CNN**

```python
# Unpack the training and test tuple
X_train = X_train.reshape(X_train.shape[0], *(28, 28, 1))
X_test = X_test.reshape(X_test.shape[0], *(28, 28, 1))
X_validate = X_validate.reshape(X_validate.shape[0], *(28, 28, 1))

print(X_train.shape)
print(y_train.shape)
print(X_validate.shape)
```

```python
def CNN_model():
  #Defining the Convolutional Neural Network
  cnn_model = Sequential()

  cnn_model.add(Conv2D(32, (3, 3), input_shape = (28,28,1), activation='relu'))
  cnn_model.add(MaxPooling2D(pool_size = (2, 2)))
  cnn_model.add(Dropout(0.25))

  cnn_model.add(Conv2D(64, (3, 3), input_shape = (28,28,1), activation='relu'))
  cnn_model.add(MaxPooling2D(pool_size = (2, 2)))
  cnn_model.add(Dropout(0.25))

  cnn_model.add(Conv2D(128, (3, 3), input_shape = (28,28,1), activation='relu'))
  cnn_model.add(MaxPooling2D(pool_size = (2, 2)))
  cnn_model.add(Dropout(0.25))

  cnn_model.add(Flatten())

  cnn_model.add(Dense(units = 512, activation = 'relu'))
  cnn_model.add(Dropout(0.5))
  cnn_model.add(Dense(units = 25, activation = 'softmax'))
  return cnn_model
```

```python
model=CNN_model()
model.compile(loss ='sparse_categorical_crossentropy', optimizer='adam' ,metrics =['accuracy'])

#Training the CNN model
history = model.fit(X_train, y_train, batch_size = 128, epochs = 20, verbose = 1,shuffle=True, validation_data = (X_validate, y_validate))

model.save("signLanguageClassificationCNN.model", save_format="h5")
```

```python
from sklearn.metrics import accuracy_score

predict_xtrain=model.predict(X_train)
predicted_train_classes=np.argmax(predict_xtrain,axis=1)
acc_score = accuracy_score(y_train, predicted_train_classes)
print('Accuracy Score of train= ',acc_score)

predict_x=model.predict(X_test)
predicted_classes=np.argmax(predict_x,axis=1)
acc_score = accuracy_score(y_test, predicted_classes)
print('Accuracy Score test = ',acc_score)

predict_xval=model.predict(X_validate)
predicted_val_classes=np.argmax(predict_xval,axis=1)
acc_score = accuracy_score(y_validate, predicted_val_classes)
print('Accuracy Score of validation = ',acc_score)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(history.history['loss'], label='Loss')
plt.plot(history.history['val_loss'], label='val_Loss')
plt.legend()
plt.grid()
plt.title('Loss evolution')

plt.subplot(2, 2, 2)
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.legend()
plt.grid()
plt.title('Accuracy evolution')
```

#**Fully Connected NN**

```python
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam

# Create a new fully connected neural network model
fc_model = Sequential()

# Flatten the input
fc_model.add(Flatten(input_shape=(28, 28, 1)))

# Add fully connected layers
fc_model.add(Dense(512, activation='relu'))
fc_model.add(Dropout(0.25))
# fc_model.add(Dense(128, activation='relu'))
# fc_model.add(Dropout(0.25))
fc_model.add(Dense(64, activation='relu'))
fc_model.add(Dropout(0.25))
fc_model.add(Dense(25, activation='softmax'))

# Compile the model
fc_model.compile(optimizer=Adam(lr=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history_fc = fc_model.fit(X_train, y_train, epochs=40, shuffle=True, validation_data=(X_validate, y_validate))

# Save the model
fc_model.save("fully_connected_model.h5")
```

```python
from sklearn.metrics import accuracy_score

predict_xtrain=fc_model.predict(X_train)
predicted_train_classes=np.argmax(predict_xtrain,axis=1)
acc_score = accuracy_score(y_train, predicted_train_classes)
print('Accuracy Score of train= ',acc_score)

predict_x=fc_model.predict(X_test)
predicted_classes=np.argmax(predict_x,axis=1)
acc_score = accuracy_score(y_test, predicted_classes)
print('Accuracy Score test = ',acc_score)

predict_xval=fc_model.predict(X_validate)
predicted_val_classes=np.argmax(predict_xval,axis=1)
acc_score = accuracy_score(y_validate, predicted_val_classes)
print('Accuracy Score of validation = ',acc_score)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(history_fc.history['loss'], label='Loss')
plt.plot(history_fc.history['val_loss'], label='val_Loss')
plt.legend()
plt.grid()
plt.title('Loss evolution')

plt.subplot(2, 2, 2)
plt.plot(history_fc.history['accuracy'], label='accuracy')
plt.plot(history_fc.history['val_accuracy'], label='val_accuracy')
plt.legend()
plt.grid()
plt.title('Accuracy evolution')
```
